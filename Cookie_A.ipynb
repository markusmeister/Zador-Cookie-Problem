```python
def plot(X, Y=None, xlabel=None, ylabel=None, legend=[], loc=None, title=None,
         xlim=None, ylim=None, xscale='linear', yscale='linear',
         xticks=None, yticks=None, xhide=False, yhide=False, yrot=False, 
         xzero=False, yzero=False, yflip=False, 
         fmts=['r-','g-','b-','y-','c-','m-'], linewidth=2, markersize=5, fillstyle='full',
         markeredgewidth=1, rasterized=False,
         grid=False, equal=False, figsize=(4,2.5), axes=None):
    """
    Plot data points.
    X: an array or list of arrays
    Y: an array or list of arrays
    If Y exists then those values are plotted vs the X values
    If Y doesn't exist the X values are plotted
    xlabel, ylabel: axis labels
    legend: list of labels for each Y series
    loc: location of the legend, like 'upper right'
    title: duh
    xlim, ylim: [low,high] list of limits for the 2 axes 
    xscale, yscale: 'linear' or 'log'
    xticks, yticks: list of locations for tick marks, or None for auto ticks
    yhide: hide the y axis?
    yrot: rotate the yaxis label to horizontal?
    yzero: zero line for the y-axis?
    fmts: a list of format strings to be applied to successive Y-series
    linewidth, markersize, fillstyle, markeredgewidth: see docs
    grid: draw a grid?
    equal: use equal aspect ratio, i.e. same scale per unit on x and y axis?
    figsize: (h,v) in inches
    rasterized: rasterize the data portion of the graph?
    axes: pre-existing axes where to draw the plot
    Returns: axes for the plot
    """
    
    if not axes: # start a new figure
        fig = plt.figure(figsize=figsize)
        axes = plt.gca()
    
    def has_one_axis(X): # Return True if X (ndarray or list) has 1 axis
        return (hasattr(X, "ndim") and X.ndim == 1 or isinstance(X, list)
                and not hasattr(X[0], "__len__"))

    if has_one_axis(X):
        X = [X]
    if Y is None:
        X, Y = [[]] * len(X), X
    elif has_one_axis(Y):
        Y = [Y]
    if len(X) != len(Y):
        X = X * len(Y)
    # axes.cla() # clears these axes
    for x, y, fmt in zip(X, Y, fmts):
        if len(x):
            axes.plot(x, y, fmt, linewidth=linewidth, markersize=markersize,
            	fillstyle=fillstyle,markeredgewidth=markeredgewidth,
            	rasterized=rasterized)
        else:
            axes.plot(y, fmt, linewidth=linewidth, markersize=markersize,
            	fillstyle=fillstyle,markeredgewidth=markeredgewidth,
            	rasterized=rasterized)
    set_axes(axes, xlabel, ylabel, legend, loc, xlim, ylim, xscale, yscale, 
             xticks, yticks, xhide, yhide, yrot, xzero, yzero, yflip, grid, equal)
    if title:
        plt.title(title)
    plt.tight_layout()

    return axes # useful if we started a new figure

def hist(X, bins=50, range=None, density=None, weights=None, histtype='stepfilled', stacked=False, 
         align='mid', color=['r','g','b','m'], alpha=0.5, linewidth=1, 
         xlabel=None, ylabel=None, legend=[], loc=None, title=None,
         xlim=None, ylim=None, xscale='linear', yscale='linear',
         xticks=None, yticks=None, xhide=False, yhide=False, yrot=False, 
         xzero=False, yzero=False, yflip=False, 
         grid=False, figsize=(5,3), axes=None, orientation='vertical'):
    """
    Histogram data.
    X: an array or list of arrays to be histogrammed
    bins: the number of bins, or a list of edges
    range: [low,high] the range over which the bins are spread
    density: normalize to unit area?
    weights: array same shape as X; each value in X only contributes that weight.
    histtype: 'bar','barstacked','step','stepfilled'
    stacked: plot multiple series on top of each other?
    align: 'left','mid',right', center the bar on left edge, center, or right edge of bin
    color: list of colors for the data series
    alpha: transparency, useful if histtype='stepfilled'
    xlabel, ylabel: axis labels
    legend: list of labels for each Y series
    loc: location of the legend, like 'upper right'
    title: duh
    xlim, ylim: [low,high] list of limits for the 2 axes 
    xscale, yscale: 'linear' or 'log'
    xticks, yticks: list of locations for tick marks, or None for auto ticks
    yhide: hide the y axis?
    yrot: rotate the yaxis label to horizontal?
    yzero: zero line for the y-axis?
    grid: draw a grid?
    figsize: (h,v) in inches
    axes: pre-existing axes where to draw the plot
    Returns: axes, n, bins, patches 
    """
    
    if not axes: # start a new figure
        fig = plt.figure(figsize=figsize)
        axes = plt.gca()
    
    def has_one_axis(X): # Return True if X (ndarray or list) has 1 axis
        return (hasattr(X, "ndim") and X.ndim == 1 or isinstance(X, list)
                and not hasattr(X[0], "__len__"))

    if has_one_axis(X):
        ndata=1
    else:
        ndata=len(X) # assuming it is an ndarray or a list of lists
    
    if len(color)<ndata: # if there aren't enough dolors provided
        col=None # use the default color sequence
    else:
        col=color[:ndata]
    
    if histtype=='step' or histtype=='stepfilled': # in these cases the legend sequence needs to be inverted for some reason
        if isinstance(legend, list):
        	legend = legend[-1::-1]
    
    n, bins, patches = axes.hist(X, bins=bins, range=range, density=density, weights=weights, 
    		histtype=histtype, stacked=stacked, align=align, 
            color=col, alpha=alpha, linewidth=linewidth,
            label=legend, orientation=orientation)

    set_axes(axes, xlabel, ylabel, legend, loc, xlim, ylim, xscale, yscale, 
    	xticks, yticks, xhide, yhide, yrot, xzero, yzero, yflip, grid, equal=False)

    if title:
        plt.title(title)
    plt.tight_layout()

    return axes, n, bins, patches # useful if we started a new figure

def set_axes(axes, xlabel, ylabel, legend, loc, xlim, ylim, xscale, yscale, 
    	xticks, yticks, xhide, yhide, yrot, xzero, yzero, yflip, grid, equal):
    """Set the axes for matplotlib."""
    axes.set_xscale(xscale)
    axes.set_yscale(yscale)
    if xlim:
        axes.set_xlim(xlim)
    else:
        axes.set_xlim(auto=True)
    if ylim:
        axes.set_ylim(ylim)
    else:
        axes.set_ylim(auto=True)
    if grid:
        axes.grid()
    if equal:
        axes.set_aspect(aspect='equal')
    if ylabel:
        if yrot:
            axes.set_ylabel(ylabel, fontsize=12, rotation=0, labelpad=15)
        else:
            axes.set_ylabel(ylabel, fontsize=12)
    if xlabel:
        axes.set_xlabel(xlabel, fontsize=12)
    axes.get_yaxis().set_visible(not yhide)
    axes.get_xaxis().set_visible(not xhide)
    if xzero:
        axes.axvline(color='black', linewidth=0.5)
    if yzero:
        axes.axhline(color='black', linewidth=0.5)
    if yflip:
        axes.invert_yaxis()
    axes.tick_params(axis = 'both', which = 'major', labelsize = 10)
    axes.tick_params(axis = 'both', which = 'minor', labelsize = 9)
    if xticks:
        axes.set_xticks(xticks,minor=False); # no minor ticks
    if yticks:
        axes.set_yticks(yticks,minor=False); # no minor ticks
    if legend:
        axes.legend(legend, loc=loc)
    plt.draw()
```

# Solution

Aug 27, 2023

We're trying to find the policy that minimizes the exected length of the search. After searching for a while in one direction, it makes sense to turn around and try the other part of the space. What is the best starting point? What is the best policy for turning around?

I tried for a while to solve this analytically, but didn't get anywhere. So here is the numerical solution using a type of Bellman equation in two variables. Conveniently it can be solved in a single pass across the 2D array.

Search space $x \in [0,H]$

Probability distribution $P(x)$

Suppose the agent has already searched the interval $[a,H-b]$. This means that

$a$ = number of points remaining to search on the left

$b$ = number of points remaining to search on the right

At this point the agent may be either at the left end of the interval, $x=a$, or at the right end, $x=H-b$. Now define two functions

$U(a,b)$ = expected time to target if the agent is at the left end, i.e. if $x=a$

$V(a,b)$ = expected time to target if the agent is at the right end, i.e. if $x=H-b$

At every step the agent decides whether to extend the searched region by a single step from its current location $x$, or to reverse across the entire searched region and continue on the other side. For example, with the agent at $x=a$, the options are to step to $x=a-1$ ($1$ step) or to reverse and go to $x=H-b+1$ ($H-b+1-a$ steps). The choice depends on what the expected time is for the two options. 

For each of these options, we have to consider that the agent might find the cookie there, in which case the search is over. For example, the probability that the cookie is at $x=a-1$, given that it wasn't found in $[a,H-b]$ is

$$\frac{P(a-1)}{1-\sum_{i=a}^{H-b}P(i)}$$

So define two functions for the probability that the cookie is *not* found in the next step, namely

$Q(a,b)=1 - \frac{P(a-1)}{1-\sum_{i=a}^{H-b}P(i)}$, for a step to $x=a-1$

$R(a,b)=1 - \frac{P(H-b+1)}{1-\sum_{i=a}^{H-b}P(i)}$, for a step to $x=H-b+1$

Then the functions $U$ and $V$ are related by

$$
U(a,b)=\min{(1+Q(a,b) U(a-1,b),H-a-b+1+R(a,b) V(a,b-1))} \\
V(a,b)=\min{(H-a-b+1+Q(a,b) U(a-1,b),1+R(a,b) V(a,b-1))}
$$

Because the values at $(a,b)$ depend only on those at $(a-1,b)$ and $(a,b-1)$, one can compute this by calculating successive diagonals where $a+b=k$, for $k=0,...,H$. 

There are obvious boundary conditions at $a=0$ and $b=0$.

Eventually this yields the values on the diagonal $U(i,H-i)$ and $V(i,H-i)$. This is the expected time with the agent at $x=i$, having searched only location $i$. Because the right end and left end of this searched interval are the same ($a=i, b=H-i$), one expects that $U(i,H-i)=V(i,H-i)$. This will be a good sanity check.

To pick the optimal starting point, one needs to take into account the possibility of finding the cookie right at the start before even taking a step. So the expected search time, starting at position $x$ is

$T(x) = (1-P(x)) U(x,H-x)$

To execute the optimal trajectory:
- Start at $x = {\rm{argmin}} \, T(x)$
- After having searched an interval $[a,b]$ decide whether to continue in the same direction or to reverse by comparing the values of $Q(a,b)U(a-1,b)$ and $R(a,b)V(a,b-1)$, adding the cost $H-a-b$ for a reversal.
- Quit when $a=b=0$.

```python
# define the arrays
Q=np.zeros((H+1,H+1))
R=np.zeros((H+1,H+1))
U=np.zeros((H+1,H+1))
V=np.zeros((H+1,H+1))
```

```python
# compute the probabilities Q and R
for i in range(1,H+1):
    for j in range(H-i+1):
        Q[i,j]=1-P[i-1]/(1-C[H-j]+C[i-1])
        
for j in range(1,H+1):
    for i in range(H-j+1):
        R[i,j]=1-P[H-j+1]/(1-C[H-j]+C[i-1]) # recall that C[-1]=0
```

# Solving for the expected time

```python
# solver function, starting from the prob distribution P
def get_TUVQR(P):
    '''
    Solves the dynamic programming problem
    Returns
    T[x]=expected search time starting from x
    U[a,b],V[a,b]=time from intermediate states during the search
    '''
    H=len(P)-1
    Q=np.zeros((H+1,H+1))
    R=np.zeros((H+1,H+1))
    U=np.zeros((H+1,H+1))
    V=np.zeros((H+1,H+1))

    C=np.cumsum(P)
    C=np.append(C,0) # this way C[-1]=0
    
    for i in range(1,H+1):
        for j in range(H-i+1):
            Q[i,j]=1-P[i-1]/(1-C[H-j]+C[i-1])
    for j in range(1,H+1):
        for i in range(H-j+1):
            R[i,j]=1-P[H-j+1]/(1-C[H-j]+C[i-1]) # recall that C[-1]=0

    U[0,0]=V[0,0]=0
    for i in range(1,H+1):
        U[i,0]=1+Q[i,0]*U[i-1,0]
        V[i,0]=H-i+1+Q[i,0]*U[i-1,0]
        for j in range(1,i):
            U[i-j,j]=min(1+Q[i-j,j]*U[i-j-1,j],H-i+1+R[i-j,j]*V[i-j,j-1])
            V[i-j,j]=min(1+R[i-j,j]*V[i-j,j-1],H-i+1+Q[i-j,j]*U[i-j-1,j])
        U[0,i]=H-i+1+R[0,i]*V[0,i-1]
        V[0,i]=1+R[0,i]*V[0,i-1]

    T=(1-P)*[U[i,H-i] for i in range(H+1)] # time from start   
    
    return T,U,V,Q,R
```

```python
# Triangular with 101 points
H=100
x=np.arange(H+1)
P=1+H/2-np.abs(x-H/2)
P/=np.sum(P)

T,U,V,Q,R=get_TUVQR(P)
plot(T);
```

# Navigating the optimal route

```python
def get_path(T,U,V,Q,R):
    '''
    Returns the optimal search path
    The traversals across previously searched space are not included
    '''
    H=len(U)-1
    x=np.argmin(T) # start at the best x
    a=x; b=H-x
    X=[]
    X+=[x]
    while a!=0 or b!=0: # finish when a=0, b=0
        if x==a: # x is on left end of searched space
            if a==0: # nothing remains on the left
                b=b-1; x=H-b # have to reverse
            elif b==0: # nothing remains on the right
                a=a-1; x=a # have to stay
            else:
                if Q[a,b]*U[a-1,b] > H-a-b + R[a,b]*V[a,b-1]: # if reversing to the other side is faster
                    b=b-1; x=H-b # reverse
                else:
                    a=a-1; x=a # stay
        else:
            if a==0: # nothing remains on the left
                b=b-1; x=H-b # have to stay
            elif b==0: # nothing remains on the right
                a=a-1; x=a # have to reverse
            else:
                if R[a,b]*V[a,b-1] > H-a-b + Q[a,b]*U[a-1,b]: # if reversing to the other side is faster
                    a=a-1; x=a # reverse
                else:
                    b=b-1; x=H-b # stay
        X+=[x]
    return X
```

```python
Y=get_path(T,U,V,Q,R)
plot(x[Y],
     xlabel='Points searched',
     ylabel='Position',
    );
```

# Meister's cookie conjectures
1. With a Gaussian distribution, the optimal search path starts at $\pi/2$.
2. With an Exponential distribution, the expected search distance is $\pi$.
